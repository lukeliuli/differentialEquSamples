### 微分对策及其应用，例子笔记: 144页，例5.3  

$$
\begin{cases}
\dot{x} = u-v \newline
x(\tau) = \xi
\end{cases}
$$

受制于：

$$
\tau \leq t \leq T_0 \newline
J(u,v) = g+ \int_{\tau}^{T_0} x^2(t) {\rm d}x ,g = 0,h = x^2(t)\newline 
0 \leq u \leq 1 \newline
0 \leq v \leq 1 \newline
\xi > 0 \newline
$$

1. 计算1
$$
H = h+\lambda f \newline
H = x^2+\lambda (u-v) \newline 
$$
2. 计算2
$$
\nu = J(u^*,v^*) \newline 
\frac{\partial \nu(\tau,\xi)}{\partial \tau}+ \frac{\partial \nu(\tau,\xi)}{\partial \xi}f(\tau) +h(\tau) = 0 \newline
f(u^*,v^*) = 0 \newline
\frac{\partial \nu(\tau,\xi)}{\partial \tau}  +h(\tau,u^0,v^0) = 0  \newline
h(\tau,u^0,v^0) =  \xi^2
$$

$$
\begin{cases}
\frac{\partial \nu(\tau,\xi)}{\partial \tau}+ \xi^2 = 0  \newline
\nu(T_0,\xi) = 0
\end{cases}
$$
3. 计算3
$$
\nu(\tau,\xi) = \xi^2(T_0-\tau) \newline
\lambda =  \frac{\partial \nu(\tau,\xi)}{\partial \xi} = 2\xi(T_0-\tau)
$$

4. 重要的公式

$$
H = h+\lambda f \newline
H = x^2+\lambda (u-v) \newline 
$$

$$
u^* = 1,v^* = 1 \newline
f(u^*,v^*) = 0  \newline
$$

$$
\begin{cases}
\frac{\partial \nu(\tau,\xi)}{\partial \tau}+ \xi^2 = 0  \newline
\nu(T_0,\xi) = 0
\end{cases}
$$

$$
\lambda =  \frac{\partial \nu(\tau,\xi)}{\partial \xi} = 2\xi(T_0-\tau)
$$

### 微分对策及其应用，例子笔记: 144页，例5.4
$$
\begin{cases}
\dot{x} = u-v \newline
x(\tau) = \xi
\end{cases}
$$

受制于：

$$
\tau \leq t \leq T_0 \newline
J(u,v) =x(T_0),g=x(T_0),h=0 \newline
-1 \leq u \leq 0 \newline
0 \leq v \leq 1 \newline
$$

1. 计算1
$$
H = h+\lambda f \newline
H = \lambda (u-v) \newline 
$$


2. 计算2
$$
u^*=0,v^*=1 \newline
f(u^*,v^*) = u^*-v^* =-1 \newline
$$

3. 计算3

$$
\frac{\partial \nu(\tau,\xi)}{\partial \tau}+ \frac{\partial \nu(\tau,\xi)}{\partial \xi}f(\tau) +h(\tau) = 0 \newline
f(\tau) = -1, h(\tau) = 0 \newline
=> \frac{\partial \nu(\tau,\xi)}{\partial \tau}+ \frac{\partial \nu(\tau,\xi)}{\partial \xi}(-1) = 0 \newline
$$


$$
\nu(T_0,\xi) = g = x(T_0) \newline
\dot{x} = u-v = -1 \newline \newline
x(\tau) = \xi \newline 
=> x = -t+C = -\tau+C = \xi => C = \xi +\tau =>x(t) = -t+ \xi +\tau  \newline
=>x(T_0) = -T_0 +\tau + \xi => \nu(T_0,\xi)  = x(T_0)  = -T_0 + \xi + \tau   \newline
$$
